```python
# Import the necessary libraries
import pandas as pd
from tabulate import tabulate
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score, f1_score
from sklearn.model_selection import train_test_split

# List of CSV files
csv_files = ["camera.csv", "laptop.csv", "mobile phone.csv", "tablet.csv", "tvs.csv", "video surveillance.csv"]
SMT = ['Camera','Laptops','Mobile phones','Tablets','TVs','Video surveillance']
# List to store the sum of reviews for each CSV file
reviews_sums = []

# Loop through each CSV file
for csv_file in csv_files:
    df = pd.read_csv(csv_file)
    reviews_sum = df["reviews"].sum()
    reviews_sums.append(reviews_sum)
```

# üåüShow and Download histogram :

```python
# Plotting the histogram
plt.figure(figsize=(10, 6))
plt.bar(SMT, reviews_sums, alpha=0.7, color="darkorange")
plt.xlabel("Products")
plt.ylabel("Number of reviews")
plt.title("Reviews Dataset")
plt.xticks()  # Rotate the x-axis labels for better readability
plt.grid(axis='y')
plt.show()
# plt.savefig('histogram.png') # show and download histogram 
```

# ‚ùÅ Results :

```python
# Create a DataFrame to store the results
results_df = pd.DataFrame({'Category': SMT, 'Reviews': reviews_sums})

# Display the results as a table
print(results_df)
```
